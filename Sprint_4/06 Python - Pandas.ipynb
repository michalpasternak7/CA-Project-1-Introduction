```python
# The Release Date is a string type, which is called object in Pandas.
# We can convert this to a date format using .to_datetime()

movies["Release Date"] = pd.to_datetime(movies["Release Date"], format = "%Y/%m/%d")
movies.info()
```

```python
# By converting to a pd.to_datetime() format we can now extract granular information from this column.

# For instance, we can create a new column called "Day of the week" and fill it with the day name of that date.

movies["Day of the week"] = movies["Release Date"].dt.day_name() # Run this code again to see what would happen if we removed the () at the end of day_name.
movies.head(5)
```

```python
# This code block aims to illustrate how we can filter the data directly.

# For instance, let's see all the movies with runtime less than 100 minutes.

runtimelessthanhundred = movies[movies["Runtime"] < 100]
```

```python
# By method chaining, we can add a df.count() to the previous code block and get the total number of movies that are less than 100 minutes.

movies[movies["Runtime"] < 100].count()
```

```python
# We can also filter by more than one condition.

# For instance, let's say we want to see all "Action" movies that were released before "1994-07-14"

movies[(movies["Release Date"] < "1994-07-14") & (movies["Genre"] == "Action")]

# The & operator in the middle means "And".
# So essentially what we are saying is filter movies "Release Date" < "1994-07-14" AND "Genre" == "Action".
```

#### 1.  Comparison Operators: These are used to compare the values on either side and decide the relation among them. They return a Boolean value (`True` or `False`).

    -   `==` Equal to
    -   `!=` Not equal to
    -   `<` Less than
    -   `>` Greater than
    -   `<=` Less than or equal to
    -   `>=` Greater than or equal to
#### 2.  Logical Operators: These are used to combine the results of comparison operators and return a single Boolean result.

    -   `&` Logical AND: True if both operands are true
    -   `|` Logical OR: True if either operand is true
    -   `~` Logical NOT: True if operand is false

```python
# Now let's say we want to filter out movies that are rated PG
# This is the syntax for that:

movies[movies["MPAA Rating"] == "PG"]

# This is called Boolean Indexing
```

```python
# We can also filter the data and save it into a new variable/df.
# There are many use cases for this and you will MAY use them in your journey.

# For instance, let's say we want to create a new df with movies that are rated "PG" only
# This is the syntax for it:

pg_movies = movies[(movies["MPAA Rating"] == "PG")]
pg_movies.head()
```
